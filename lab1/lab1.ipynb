Wykres funkcji $f(x) = {x}^2 + 5$ 
1. dla $x>-1$ i $x<1$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

#wykresy funckji dla trzech poszczegÃ³lnych wartoÅ›ci x

f_x = lambda x: x**2 + 5
x_1 = np.linspace(-1, 1, 1000)
y_1 = f_x(x_1)

plt.plot(x_1, y_1, "b", label = "x^2 + 5")
plt.title("Wykres funkcji dla x>-1 x<1")
plt.legend()
plt.ylabel("y(x)")
plt.xlabel("x")
plt.grid()
plt.show()
```

2. dla $x>-6$ i $x<6$

```python
x_2 = np.linspace(-6, 6, 1000)
y_2 = f_x(x_2)

plt.plot(x_2, y_2, "r", label = "x^2 + 5")
plt.title("Wykres funkcji dla x>-6 x<6")
plt.legend()
plt.ylabel("y(x)")
plt.xlabel("x")
plt.grid()
plt.show()
```

2. dla $x>0$ i $x<5$

```python
x_3 = np.linspace(0, 5, 1000)
y_3 = f_x(x_3)

plt.plot(x_3, y_3, "g", label = "x^2 + 5")
plt.title("Wykres funkcji dla x>0 x<5")
plt.legend()
plt.ylabel("y(x)")
plt.xlabel("x")
plt.grid()
plt.show()
```
