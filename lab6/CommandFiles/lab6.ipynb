```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bins_labels(bins):
    bin_w = (max(bins) - min(bins)) / (len(bins) - 1)
    plt.xticks(np.arange(min(bins)+bin_w/2, max(bins)+bin_w/2+0.001, bin_w), bins)
    plt.xlim(bins[0], bins[-1])
```

```python
data = pd.read_csv("../OriginalData/wojewodztwo.csv")

time_of_having = data[["Dni od zakupu"]]
time_of_having.to_csv("../AnalysisData/time_of_having.csv")

bins = np.arange(0,20,1)
ax = time_of_having.plot(kind="hist",bins = bins, legend = False)
ax.set_ylabel("Number of bougth products")
ax.set_xlabel("Number of days before giving review")
ax.set_title("Number of days before giving review")
bins_labels(bins)
fig = ax.get_figure()
fig.savefig("../Documents/time_of_having.jpg")

brand_purchases = data[["Marka","Wiek kupujcego"]]
brand_purchases = brand_purchases.groupby("Marka").count()
brand_purchases.to_csv("../AnalysisData/brand.csv")

ax = brand_purchases.plot(kind= "bar", legend = False)
ax.set_ylabel("Number of sold products")
ax.set_xlabel("Crand of the vacuum")
for i in ax.patches:
    ax.annotate(format(i.get_height(), 'd'),
                   (i.get_x() + i.get_width() / 2,
                    i.get_height()))
ax.set_title("Number of sold vacuums by brand")
fig = ax.get_figure()
fig.savefig("../Documents/brand.jpg")

customer_age = data[["Wiek kupujcego"]]
customer_age.to_csv("../AnalysisData/age.csv")

bins = np.arange(15,80,5)
ax = customer_age.plot(kind= "hist", legend = False)
ax.set_title("Age of buyer")
ax.set_xlabel("Age range")
ax.set_ylabel("Number of buyers")
bins_labels(bins)
fig = ax.get_figure()
fig.savefig("../Documents/age.jpg")

gender_count = data[["Pe kupujcego","Unnamed: 0"]]
gender_count = gender_count.groupby("Pe kupujcego").count()
gender_count.to_csv("../AnalysisData/gender.csv")

ax = gender_count.plot(kind= "bar", legend = False)
ax.set_ylabel("Number of sold vacuums")
ax.set_xlabel("Gender od consumer")
for i in ax.patches:
    ax.annotate(format(i.get_height(), 'd'),
                   (i.get_x() + i.get_width() / 2,
                    i.get_height()))
ax.set_title("Number of sold products by gender")
fig = ax.get_figure()
fig.savefig("../Documents/gender.jpg")

rates = data[["Ocena"]]
rates.to_csv("../AnalysisData/rate.csv")

bins = np.arange(0,6,0.5)
ax = rates.plot(kind= "hist", legend = False)
ax.set_title("Review of products")
ax.set_xlabel("Rate")
ax.set_ylabel("number of given reviews")
bins_labels(bins)
fig = ax.get_figure()
fig.savefig("../Documents/review.jpg")
```
